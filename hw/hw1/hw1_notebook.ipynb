```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi'] = 200
plt.style.use('fivethirtyeight')
from matplotlib.figure import figaspect
```

```python
d = [0.016119323, 0.003255398, 0.003580815, 0.009251528, 0.051583417, 0.289241385, 2.962166883, 60.00520416, 60.002103951, 60.002885675, 60.002089819, 60.020584561]
w, h = figaspect(1/4)
fig, ax = plt.subplots(figsize=(w,h))
plt.plot(list(range(4, 16)), d, marker='o')
plt.ylabel('Running Time (sec)')
plt.xlabel('Number of Pigeons')
plt.title('Runtime of the Pigeon-Hole Problem using MiniSAT (timeout at 60 seconds)')
plt.tight_layout()
plt.savefig("figs/pigeon_hole_runtime.pdf", bbox_inches='tight')
```

```python
w, h = figaspect(1/4)
fig, ax = plt.subplots(figsize=(w,h))
plt.semilogy(list(range(4, 16)), d, marker='o')
plt.ylabel('Log(Running Time) (sec)')
plt.xlabel('Number of Pigeons')
plt.title('Runtime of the Pigeon-Hole Problem using MiniSAT (timeout at 60 seconds)')
plt.tight_layout()
plt.savefig("figs/pigeon_hole_runtime_log.pdf", bbox_inches='tight')
```
